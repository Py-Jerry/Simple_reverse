```python
# 首先导入需要使用的工具包
import requests  # 发送请求
import re  # 用于解析出文本数据中的电话号码数据
import pandas as pd # 导入pandas模块用于存储数据
```

```python
# 定义空的列表用于存储数据
data = []
```

```python
def extract_phone(text: str) -> str | None:
    """
    从文本中提取第一个电话号码（格式如 347-702-2746）。
    支持不同分隔符，如 -, ., 空格。
    若未找到则返回 None。
    """
    pattern = re.compile(r'\b\d{3}[-.\s]\d{3}[-.\s]\d{4}\b')
    match = pattern.search(text)
    return match.group() if match else None
```

```python
def get_dataapi(page: int | str):
    """
        此函数用于发送请求至会返回数据的api
    :param page: 页码，更换此参数实现换页
    :return: 
    """
    # 身份信息
    cookies = {
        '_forum_session': 'G2jT%2BVwmo74UZTGe6MVzhoR%2FxaaTbhujtBuPFa8oeVHn3OVMFA07qNOZRsxNRkgJEBnt8Cet4XisFlSHwEXXkb%2Fv%2BuTonQXCzEIAlNx40wqBjakj7KSuhgtQZy7%2Ffjma95oKbiefoY0yfBlIgkn3tWGmeW3jORb0dWz%2F7GMyCoIarg%3D%3D--39JUjZrHphPRzY8P--gM1o%2FEg3LU8RV0wPWlhUKw%3D%3D',
    }
    # 请求头部信息
    headers = {
        'referer': 'https://huarenzhaopin.com/',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36',
    }
    print(f'=======================>正在获取第{page}页数据...')
    # 传入参数
    params = {
        'no_definitions': 'true',
        'page': str(page),  # 这里是页码，通过修改此参数的值来获取不同页的数据 -> max_page:197
    }
    try:    # 捕获异常
        response = requests.get('http://huarenzhaopin.com/latest.json', params=params, cookies=cookies, headers=headers)
    except requests.exceptions.SSLError:    # 捕获网络异常
        return get_dataapi(page)    # 捕获成功则重新请求
    if response.status_code != 200: # 检查请求码是否为200
        print(f'请求失败，状态码:{response.status_code}')
        return get_dataapi(page)    # 不是则重新请求
    topics = response.json()['topic_list']['topics']

    for topic in topics:
        fancy_title = topic['fancy_title']
        telephone_number = extract_phone(fancy_title)
        if telephone_number is None:
            continue
        data.append(telephone_number.replace(' ',''))
        print(f'标题:{fancy_title} ============= 电话:{telephone_number}')
```

```python
if __name__ == '__main__':
    for page in range(1, 198):
        get_dataapi(page)
    # 进行数据去重
    data = list(set(data))
    print(f'数据获取完毕，共有{len(data)}条数据')
    # 存储到excel
    df = pd.DataFrame(data, columns=['电话'])
    df.to_excel('huarenzhaopin——telephone_number.xlsx', index=False)
    print('数据保存完毕')
```
